## Data Consistency Checks

### Handling Missing Values

```python
# Checking missing values
toronto_listings.isnull().sum()
```

**Important Notes:** 
1. Neighbourhood column has 45% missing data. This column will be dropped.
2. neighbourhood_group_cleansed has most of the data missing, and therefore we will remove this column from analysis.
3. The columns with less than 5% of the missing data will be kept as it is as the number of missing data isn't significant. For the numerical columns with more than 5% missing data, the data will be imputed if required for the analysis.
4. The price column has 25% of the data missing. This data can be imputed by mean by choosing appropriate features. However, since we are predict price with the help of features that are correlated to it, we will be removing the missing price entries. Keep in mind this may create bias in the data set and the results may not be accurate.

```python
# Removing missng prices
toronto_listings = toronto_listings[~toronto_listings['price'].isnull()]
```

```python
toronto_listings.isnull().sum()
```

We will now drop neighbourhood, neighbourhood_group_cleansed columns.

```python
toronto_listings = toronto_listings.drop(columns = ['neighbourhood','neighbourhood_group_cleansed'])
```

```python
#Converting host_acceptance_rate to float type after removing % sign
toronto_listings['host_acceptance_rate'] = toronto_listings['host_acceptance_rate'].replace({'\%':''},regex=True).astype('float')
```

```python
# Imputing mean value for acceptance rate
toronto_listings.loc[toronto_listings['host_acceptance_rate'].isnull(),'host_acceptance_rate'] = toronto_listings['host_acceptance_rate'].mean()
```

```python
# Concerting host response rate to float type after removing % sign
toronto_listings['host_response_rate'] = toronto_listings['host_response_rate'].replace({'\%':''},regex=True).astype('float')
```

```python
# Imputing mean value for host_response_rate 
toronto_listings.loc[toronto_listings['host_response_rate'].isnull(),'host_response_rate'] = toronto_listings['host_response_rate'].mean()
```

```python
# Imputing mean value for reviews_per_month
toronto_listings.loc[toronto_listings['reviews_per_month'].isnull(),'reviews_per_month'] = toronto_listings['reviews_per_month'].mean()
```

```python
# Removing host response time
toronto_listings = toronto_listings.drop(columns = 'host_response_time')
```
